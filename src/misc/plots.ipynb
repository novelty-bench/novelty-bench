```python
%cd ~/src/second-best-bench/
import json

import altair as alt
import pandas as pd


def palatino():
    font = "TeX Gyre Pagella"

    return {
        "config": {
            "title": {"font": font},
            "axis": {"labelFont": font, "titleFont": font},
            "header": {"labelFont": font, "titleFont": font},
            "legend": {"labelFont": font, "titleFont": font},
        }
    }


alt.themes.register("palatino", palatino)
alt.themes.enable("palatino")
```

```python
equiv_counts = []

for model, _model_alias in {
    "gpt-3.5-turbo": "GPT-3.5-turbo",
    "gpt-4o": "GPT-4o",
    "gpt-4o-mini": "GPT-4o mini",
}.items():
    with open(f"evals/{model}/summary.json") as f:
        data = json.load(f)
    for k, freq in enumerate(data["equivalence-classes-histogram"], 1):
        equiv_counts.append(
            {
                "model": model,
                "k": k,
                "freq": freq,
            }
        )

equiv_df = pd.DataFrame(equiv_counts)

# Create quality at k visualization
freq_plot = (
    alt.Chart(equiv_df)
    .mark_line(point=True)
    .encode(
        x=alt.X("k:O", title="k"),
        y=alt.Y("freq:Q", title="Frequency"),
        color=alt.Color("model:N", legend=alt.Legend(title="Model")),
    )
    .properties(width=300, height=300, title="Distribution of equivalence classes")
)
freq_plot.save("plots/equiv-classes.json")
freq_plot
```

```python
# Prepare data for quality at k
quality_data = []

for model, _model_alias in {
    "gpt-3.5-turbo": "GPT-3.5-turbo",
    "gpt-4o": "GPT-4o",
    "gpt-4o-mini": "GPT-4o mini",
}.items():
    with open(f"evals/{model}/summary.json") as f:
        data = json.load(f)
    for k, quality in enumerate(data["quality-at-k"], 1):
        quality_data.append(
            {
                "model": model,
                "k": k,
                "quality": quality,
            }
        )

quality_df = pd.DataFrame(quality_data)

# Create quality at k visualization
quality_plot = (
    alt.Chart(quality_df)
    .mark_line(point=True)
    .encode(
        x=alt.X("k:O", title="k"),
        y=alt.Y("quality:Q", title="Score"),
        color=alt.Color("model:N", legend=alt.Legend(title="Model")),
    )
    .properties(width=300, height=300, title="Quality@k")
)

quality_plot.save("plots/quality@k.json")
quality_plot
```
